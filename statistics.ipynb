```python
import json
import pandas as pd

with open("hugi/data.json", "r") as f:
    data_hugi = json.load(f)

with open("bland/data.json", "r") as f:
    data_bland = json.load(f)

hugi = pd.DataFrame(data_hugi)
bland = pd.DataFrame(data_bland)
```

```python
hugi.head()
```

```python
import matplotlib.pyplot as plt

dataset = [("Hugi.is ", hugi), ("bland.is ", bland)]

for name, data in dataset:
    amount_of_responses = [len(x) for x in data["responses"]]
    plt.hist(amount_of_responses, bins=range(0, 20))
    plt.title(name + "frequency of responses")
    plt.xlabel("Number of responses")
    plt.ylabel("Frequency")
    plt.show()

for name, data in dataset:
    data = data[data["responses"].apply(lambda x: len(x) > 0)] # remove empty responses
    print(data.shape)
    length_of_response = len(data["responses"][0])
    plt.hist(length_of_response, bins=range(0, 20))
    plt.title(name + "length of responses")
    plt.xlabel("Length of responses")
    plt.ylabel("Frequency")
    plt.show()
```
